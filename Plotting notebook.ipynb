```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = 'cifar'
model = 'vgg'

PATH = 'output/'+dataset+'_'+model+'.json'
```

```python
df = pd.read_json(PATH)# first number is the experiment number
                       # second number is the number epoch
```

```python
plt.figure(figsize = (8,8))

for i in range(len(df)): #experiment number
    val_loss = np.empty(0)
    
    name = df.config[i]['opt']['name']
    if 'beta' in df.config[i]['opt'].keys():
        beta = str(df.config[i]['opt']['beta'])
        
    else:
        beta = str(0.9)
    lr = 'lr =' + str(df.config[i]['opt']['lr'])
    for j in range(len(df.history[i])): #epoch number
        val_loss = np.append(val_loss, df.history[i][j]['val_score'])
    plt.plot(range(len(df.history[i])),val_loss, label = name +' '+ lr + ' ' + 'beta =' + beta)
plt.xlabel('number of epochs')
plt.ylabel('validation accuracy')
plt.title(dataset+' + ' + model + ' + no weight decay')
plt.grid()
plt.legend()
    
```
